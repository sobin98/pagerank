# random teleport vs topic specific teleport



```python
import pickle
with open('data_biased.pkl','rb') as f:
    biasd= pickle.load(f)
```

```python
with open('data_naive.pkl','rb') as f:
    naive= pickle.load(f)
```

```python
biased_x=list(range(1,30000))
biased_y=[x[1] for x in biasd]
naive_x=list(range(1,30000))
naive_y=[x[1] for x in naive]
```

```python
import matplotlib.pyplot as plt
plt.bar(naive_x,naive_y)

plt.title("before bias", size=15)

plt.xlim([0, 30000])  
plt.ylim([0, 0.001])

plt.xlabel('Page ID')
plt.ylabel('page rank')

plt.show()
```

```python
plt.bar(biased_x,biased_y)
plt.title("before bias", size=15)
plt.xlim([0, 30000])  
plt.ylim([0, 0.005])  

plt.xlabel('Page ID')
plt.ylabel('page rank')

plt.show()
```

bias 적용전에는 초반부의 페이지들에 페이지 랭크값이 집중되어있었지만

bias 적용후에 개인화된 페이지 랭크값이 적용되어서 다양하게 분포되어 있음을 확인할 수 있다.

```python
import numpy as np
naive_y=np.array(naive_y)*6
```

```python
plt.bar(biased_x,biased_y,color="#dd0000")
plt.bar(naive_x,naive_y,color="#ddff00")

plt.xlim([0, 30000])  
plt.ylim([0, 0.005])  

plt.xlabel('Page ID')
plt.ylabel('page rank')

plt.show()
```

위 막대그래프는 bias 적용전과 후를 한눈에 보기 위해서 하나의 그래프로 표현한 것
